## failed된 데이터 가져오기
이때 failed된 데이터중에서 결측치가 있거나, 없는 날짜가 있으면 보간법으로 값을 채우겠다.<br>
여기서 마지막으로 생성된 selected_filtered_data_failed는 1월1일부터 6월 30일까지 데이터가 다 존재하는 failed disk_id들만 모은것이고<br>
failed disk_id 는 304개 였지만 모든 날짜가 다 있는 disk_id는 139개이다.그리고 보간하기전에 모든 날짜가 있는 disk_id는 55개이다.

```python
failed_data=data[data['label']==1]
unique_ids = failed_data['disk_id'].unique()
```

```python
len(unique_ids)
```

```python
unique_disk_ids = []  # 유일한 disk_id를 저장할 리스트

for disk_id in unique_ids:
    disk_data = data[data['disk_id'] == disk_id]
    if len(disk_data) == 180:  # 행의 개수가 180인 경우에만 추가
        unique_disk_ids.append(disk_id)

unique_disk_ids = pd.Series(unique_disk_ids, name='disk_id') # 약 3초걸림
```

보간하기 전에 180개의 행을 가지고 있는 disk_id를 보자

```python
unique_disk_ids
```

fail된 disk중에서 180개가 안되는것들을 보간해보자

```python
dfs_to_concat = []  # 추가할 데이터프레임들을 저장할 리스트

for disk_id in unique_ids:
    disk_data = data[data['disk_id'] == disk_id].copy()
    disk_data['ds'] = pd.to_datetime(disk_data['ds'], format='%Y%m%d')
    disk_data = disk_data.sort_values(by='ds')
    disk_data = disk_data.set_index('ds', drop=True)
    all_dates = pd.date_range(start=disk_data.index.min(), end=disk_data.index.max(), freq='D')
    disk_data = disk_data.reindex(all_dates)
    disk_data = disk_data.interpolate(method='linear')
    dfs_to_concat.append(disk_data)

filled_data_failed = pd.concat(dfs_to_concat)  # 데이터프레임들을 연결하여 새로운 데이터프레임 생성

filled_data_failed = filled_data_failed.reset_index()  # 인덱스 재설정
```

```python
filled_data_failed
```

1월1일부터 시작안하는 failed_disk들도 많아서  304*181 = 55024로 딱 떨어지지는 않는다.<br>
그래서 181개로 보간되는 것들을 살펴보니 138개가 있어서 failed한 데이터 138개 healthy한 데이터 138개를 가져와서 비율을 맞추겠다.

```python
len(unique_ids)
```

```python
unique_disk_ids = []  # 유일한 disk_id를 저장할 리스트

for disk_id in unique_ids:
    disk_data = filled_data_failed[filled_data_failed['disk_id'] == disk_id]
    if len(disk_data) == 181:  # 행의 개수가 180인 경우에만 추가
        unique_disk_ids.append(disk_id)

unique_disk_ids = pd.Series(unique_disk_ids, name='disk_id') 
```

## failed된 disk_id만큼 healthy한 disk_id가져오기
failed된 디스크중에서 모든 날짜가 다 존재하는 disk_id는 139개여서 healthy한 디스크 아이디 중에서 139개를 가져와서 데이터를 만들겠다.
결측치가 없는 데이터중에서 random으로 가져오겠다. 

```python
healthy_data=data[data['label']==0]
unique_ids = healthy_data['disk_id'].unique()
```

```python
unique_ids
```

```python
unique_disk_ids = []  # 유일한 disk_id를 저장할 리스트

for disk_id in unique_ids:
    disk_data = healthy_data[healthy_data['disk_id'] == disk_id]
    if len(disk_data) == 180:  # 행의 개수가 180인 경우에만 추가
        unique_disk_ids.append(disk_id)

unique_disk_ids = pd.Series(unique_disk_ids, name='disk_id') # 약 6분걸림
```

```python
len(unique_disk_ids)
```

```python
unique_disk_ids
```

```python
import random

random.seed(42)  # 랜덤 선택 결과 재현을 위한 시드 설정

selected_ids = random.sample(list(unique_disk_ids), 139)  # unique_disk_ids에서 랜덤으로 304개 선택

selected_disk_ids = pd.Series(selected_ids, name='disk_id')
```

```python
selected_disk_ids
```

근데 어이없게도 healthy한 데이터중에서 selected_disk_ids를 뽑았는데 이 데이터에서도 6월 13일이 없다. 애초에 모든 데이터가 6월 13일이 없다 그래서 6월 13일은 넣고 181일로 맞춰주는 작업을 하고 이때 보간법을 이용하겠다.

```python
dfs_to_concat = []  # 추가할 데이터프레임들을 저장할 리스트

for disk_id in selected_disk_ids:
    disk_data = data[data['disk_id'] == disk_id].copy()
    disk_data['ds'] = pd.to_datetime(disk_data['ds'], format='%Y%m%d')
    disk_data = disk_data.sort_values(by='ds')
    disk_data = disk_data.set_index('ds', drop=True)
    all_dates = pd.date_range(start=disk_data.index.min(), end=disk_data.index.max(), freq='D')
    disk_data = disk_data.reindex(all_dates)
    disk_data = disk_data.interpolate(method='linear')
    dfs_to_concat.append(disk_data)

selected_filtered_data_healthy = pd.concat(dfs_to_concat)  # 데이터프레임들을 연결하여 새로운 데이터프레임 생성

selected_filtered_data_healthy = selected_filtered_data_healthy.reset_index()  # 인덱스 재설정
```
